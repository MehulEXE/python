```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.__version__
```

```python
df=pd.read_csv(r"C:\Users\dell\OneDrive\Documents\GitHub\python\spotify.csv",index_col='Date',parse_dates=True)
```

```python
#index of the data will be labled in x-axis
plt.figure(figsize=(16,9))
sns.lineplot(data=df)
# for only to plot any specific data  mention the name ['name'] after df add ,label='something'
plt.title("spotify records")
plt.xlabel("listens per million")
plt.ylabel('dates')
plt.legend()
# plt.grid(True,alpha=0.5)
plt.show()
```

```python
'''BAR PLOT '''
#reauired x and y seprately
plt.figure(figsize=(12,4))
sns.barplot(x=df.index,y=df['HUMBLE.'])
```

```python
'''HEATMAP PLOT'''
#Required only the data it will plot accordingly
data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68, 24.59],
    'tip': [1.01, 1.66, 3.50, 3.31, 3.61],
    'sex': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'smoker': ['No', 'No', 'No', 'No', 'No'],
    'day': ['Sun', 'Sun', 'Sun', 'Sun', 'Sun'],
    'time': ['Dinner', 'Dinner', 'Dinner', 'Dinner', 'Dinner'],
    'size': [2, 3, 3, 2, 4]
}
df = pd.DataFrame(data)
cc=df.corr(numeric_only=True)#corr only selects the numeric data
sns.heatmap(data=cc,annot=True,cmap='jet')
```
